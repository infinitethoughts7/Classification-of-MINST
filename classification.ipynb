## Ceating Train & Test Datasets

```python
x_train , x_test, y_train, y_test = x[:60000], x[60000:], y[:60000], y[60000:]
```

## Training Binary Classification ( SGD )

```python
positive_class = 2
y_train_5 = (y_train == positive_class) # True for all 5s, False for other digits.
y_test_5 = (y_test == 2)
```

```python
#some_digit = np.reshape(some_dig, (-1, 1))
print(some_dig)
some_dig.shape
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=7)
sgd_clf.fit(x_train, y_train_5) #fit the model
sgd_clf.predict([some_dig]) #Preditction

```

## Perfomance Measure 

```python
from sklearn.model_selection import cross_val_score
cross_val_score(sgd_clf, x_train , y_train_5, cv =5 , scoring= "accuracy")
```

## Normal base Classifier 

```python
from sklearn.base import BaseEstimator
class never5Classifier(BaseEstimator):
    def fit(self, x, y= None):
        pass
    def predict(self, x):
        return np.zeros((len(x), 1), datatype = bool)
    
never_5_clf = never5Classifier()
cross_val_score(never_5_clf , x_train, y_train_5 , cv = 3 , scoring= "accuracy")

'''That’s right, it has over 90% accuracy! 
This is simply because only about 10% of the images are 5s, 
so if you always guess that an image is not a 5, you will be 
right about 90% of the time. Beats Nostradamus.'''
```

```python
from sklearn.model_selection import cross_val_predict
y_train_pred = cross_val_predict(sgd_clf, x_train , y_train_5, cv = 3)
```
